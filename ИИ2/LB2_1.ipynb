**Переменные таблицы:**

**Количество посещенных врачей - Number_of_Doctors_Visited - общее количество разных врачей, которых посетил пациент = { 1: 0-1 врачей, 2: 2-3 врача, 3: 4 или более врачей }**

**Возраст - Age - Возрастная группа пациента = { 1: 50-64 2: 65-80 }** 

**Физическое здоровье - Physical_Health - самооценка физического благополучия пациента = { -1: Отказано 1: Отлично 2: Очень хорошо 3: Хорошо 4: Справедливо 5: Плохо }** 

**Психическое здоровье -Mental_Health - самооценка психического или психологического здоровья пациента = { -1: Отказано 1: Отлично 2: Очень хорошо 3: Хорошо 4: Справедливо 5: Плохо }**

**Здоровье зубов - Dental_Health - самооценка состояния полости рта пациента = { -1: Отказано 1: Отлично 2: Очень хорошо 3: Хорошо 4: Справедливо 5: Плохо }**

**Занятость - Employment - Статус занятости пациента или информация, связанная с работой = { -1: Отказано 6 1: Работает полный рабочий день 2: Работает неполный рабочий день 3: На пенсии 4: В данный момент не работает }**

**Стресс мешает пациенту спать - Stress_Keeps_Patient_from_Sleeping	- влияет ли стресс на способность пациента спать = { 0: Нет 1: Да }**

**Лекарства мешают пациенту спать - Medication_Keeps_Patient_from_Sleeping - влияют ли лекарства на сон пациента = { 0: Нет 1: Да }**

**Боль мешает пациенту спать - Pain_Keeps_Patient_from_Sleeping - Нарушает ли физическая боль сон пациента = { 0: Нет 1: Да }**

**Потребность в туалете мешает пациенту спать - Bathroom_Needs_Keeps_Patient_from_Sleeping - влияет ли необходимость посещения туалета на сон пациента = { 0: Нет 1: Да }**

**Неизвестно Мешает пациенту спать - Uknown_Keeps_Patient_from_Sleeping - неустановленные факторы, влияющие на сон пациента = { 0: Нет 1: Да }**

**Проблемы со сном - Trouble_Sleeping - общие проблемы или трудности, с которыми сталкивается пациент со сном = { 0: Нет 1: Да }**

**Рецептурные снотворные препараты - Prescription_Sleep_Medication - Информация о любых снотворных препаратах, назначенных пациенту = { -1: Отказано 1: Использовать регулярно 2: Использовать изредка 3: Не использовать }**

**Расовая принадлежность - Race - Пациент расовое или этническое происхождение = { -2: Не спрашивалось -1: ОТКАЗАНО 1: Белый, неиспаноязычный 2: Черный, неиспаноязычный 3: Другой, неиспаноязычный 4: испаноязычный 5: 2+ расы, неиспаноязычные }**

**Пол - Gender - Гендерная идентичность пациента = { -2: Не спрашивалось -1: ОТКАЗАНО 1: Мужчина 2: Женщина }**

```python
# Импорт необходимых библиотек
import pandas as pd  # Для работы с таблицами данных (DataFrame)
import matplotlib.pyplot as plt  # Для создания визуализаций
import seaborn as sns  # Для построения улучшенных графиков на основе matplotlib

# 1. Загрузка данных
try:
    data = pd.read_csv("NPHA-doctor-visits.csv")  
    print("Данные успешно загружены!") 
except FileNotFoundError:  # Обработка ошибки, если файл не найден
    print("Файл NPHA-doctor-visits.csv не найден.")
    exit()  
except Exception as e:  # Общая обработка других возможных ошибок
    print(f"Ошибка при загрузке: {e}")  
    exit()  

# 2. Первичный анализ данных
print(f"Информация о данных:\n{data.info()}")  # Выводим информацию о структуре набора данных (типы столбцов, пропуски)
print(f"Размер данных: {data.shape}")  # Отображаем размеры таблицы (число строк и столбцов)
print(f"Названия столбцов: {data.columns}")  # Выводим список всех имен столбцов

# 3. Проверка пропусков
missing_values = data.isnull().sum()  # Подсчитываем количество пропусков (NaN) в каждом столбце
if missing_values.any():  # Проверяем, есть ли пропуски в наборе данных
    print(f"Пропущенные значения: {missing_values[missing_values > 0]}")  # Выводим информацию о столбцах с пропусками
    data.fillna(0, inplace=True)  # Заполняем пропуски значением 0
    print("Пропуски заполнены нулями.") 

# 4. Сравнение двух признаков с помощью гистограммы
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns  # Выбираем числовые столбцы из данных
if len(numeric_columns) >= 2:  # Проверяем, достаточно ли числовых столбцов для анализа
    feature_x = numeric_columns[2]  # Выбираем третий числовой признак для анализа
    feature_y = numeric_columns[3]  # Выбираем четвертый числовой признак для анализа

    plt.figure(figsize=(10, 6))  # Устанавливаем размер графика
    plt.hist(data[feature_x], bins=20, alpha=0.5, label=feature_x, color='blue')  # Строим гистограмму для первого признака
    plt.hist(data[feature_y], bins=20, alpha=0.5, label=feature_y, color='orange')  # Строим гистограмму для второго признака

    plt.title(f"Сравнение распределений {feature_x} и {feature_y}", fontsize=16)  # Заголовок графика
    plt.xlabel("Значение")  # Подпись оси X
    plt.ylabel("Частота")  # Подпись оси Y
    plt.legend(loc='best', fontsize=12)  # Добавляем легенду
    plt.tight_layout()  # Оптимизируем расположение элементов на графике
    plt.show()  # Отображаем график
else:
    print("Недостаточно числовых признаков для гистограммы.")  

# 5. Проверка распределений для Number of Doctors Visited и Age
print(data['Number of Doctors Visited'].value_counts())  # Подсчитываем частоту уникальных значений для признака "Number of Doctors Visited"
print(data['Age'].describe())  # Выводим статистическое описание признака "Age" (среднее, медиана, квартили)

# 6. Сравнение двух признаков (если есть два числовых признака)
if len(numeric_columns) >= 2:  # Проверяем, достаточно ли числовых признаков для анализа
    feature_x = numeric_columns[0]  # Выбираем первый числовой признак
    feature_y = numeric_columns[1]  # Выбираем второй числовой признак
    plt.figure(figsize=(8, 6))  # Устанавливаем размер графика
    sns.scatterplot(x=data[feature_x], y=data[feature_y], alpha=0.7)  # Строим диаграмму рассеяния (scatter plot)
    plt.title(f"Сравнение {feature_x} и {feature_y}", fontsize=16)  # Заголовок графика
    plt.xlabel(feature_x)  # Подпись оси X
    plt.ylabel(feature_y)  # Подпись оси Y
    plt.show()  # Отображаем график
else:
    print("Недостаточно числовых признаков для графика.")
```

**Выводы**

**График показывает слабую положительную связь между физическим и ментальным здоровьем, от сюда следует, что при улучшении одного показателя другой также имеет тенденцию к улучшению.**

**Сравнение Кол-ва посещений врачей и Возраста пациентов**

**На графике видно 3 точки, Все они находятся на уровне 2 по оси х, т.е. это информация касательно людей возрастом 65 - 80 лет. Что же по оси у, виднеются значения 1, 2 и 3, т.е. информация касается посещения врачей 1 (0-1 врачей), 2 (2-3 врача) и 3 (4 и более врача)**
