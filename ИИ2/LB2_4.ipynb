**В качестве набора ресурсов, буду использовать cirrhosis.csv**

**Описание набора:**

**Цирроз возникает в результате длительного повреждения печени, приводящего к обширному рубцеванию, часто из-за таких заболеваний, как гепатит или хроническое употребление алкоголя.**

**Приведённые данные взяты из исследования первичного билиарного цирроза (ПБЦ) печени, проведённого в клинике Майо с 1974 по 1984 год. Используйте 17 клинических признаков для прогнозирования состояния выживаемости пациентов с циррозом печени. Состояния выживаемости включают 0 = D (смерть), 1 = C (выбыл из исследования), 2 = CL (выбыл из исследования из-за трансплантации печени).**

**Перечень признаков:**

**1. ID - уникальный идентификатор**

**2. N_Days - количество дней между регистрацией и более ранней датой смерти, трансплантации или анализа в июле 1986 года**

**3. Status - состояние пациента C (цензурировано), CL (цензурировано из-за повреждения печени) или D (смерть)**

**4. Drug - тип препарата D-пеницилламин или плацебо**

**5. Age - возраст в [днях]**

**6. Sex - M (мужчина) или F (женщина)**

**7. Ascites наличие асцита N (Нет) или Y (Да)**

**8. Hepatomegaly - наличие гепатомегалии N (Нет) или Y (Да)**

**9. Spiders - наличие количество пауков N (Нет) или Y (Да)**

**10. Edema - наличие отека N (нет отека и нет диуретической терапии отека), S (отек присутствует без диуретиков или отек устраняется диуретиками) или Y (отек, несмотря на диуретическую терапию)**

**11. Bilirubin - уровень билирубина в сыворотке крови в [мг/дл]**

**12. Cholesterol - уровень холестерина в сыворотке крови в [мг/дл]**

**13. Albumin - альбумин в [г / дл]**

**14. Copper - медь в моче в [мкг / день]**

**15. Alk_Phos - щелочная фосфатаза в [ед / л]**

**16. SGOT: SGOT (сывороточная глутаминово-оксалоуксусная трансаминаза) — фермент, обнаруживаемый в клетках печени и сердца. в [Ед / мл]**

**17. Triglycerides - триглицериды в [мг / дл]**

**18. Platelets - количество тромбоцитов на кубический объем [мл / 1000]**

**19. Prothrombin - протромбиновое время в секундах [с]**

**20. Stage - гистологическая стадия заболевания (1, 2, 3 или 4)**

**Для выполнения 4 лабораторной работы необходимо:**

Выполните построение модели классификации на основе дерева 
классификации. В ходе решения задачи необходимо решить следующие 
подзадачи:

2.1 Построение логического классификатора с заданием max_depth
(максимальной глубины) и max_features (максимального количества 
признаков) пользователем (установить любые); визуализация дерева решений 
для выбранных исследователем параметров (в формате .png)

2.2 Вычисление оценки cross validation (MSE) для различнх значений 
max_depth (построить график зависимости);

2.3 Вычисление оценки cross validation (MSE) для различнх значений 
max_features (построить график зависимости);

2.4 Вычислите оптимальные значения max_depth и max_features. 
Обоснуйте свой выбор. Продемонстрируйте использование полученного 
классификатора. 

2.5 Выведите дерево в формате .png;

2.6 Выведите решающие границы полученной модели.


```python
# Импорт необходимых библиотек
import pandas as pd  # Для работы с таблицами данных (DataFrame)
from sklearn.model_selection import train_test_split, cross_val_score  # Для разделения данных и кросс-валидации
from sklearn.tree import DecisionTreeClassifier, plot_tree  # Для построения и визуализации дерева решений
from sklearn.preprocessing import LabelEncoder  # Для кодирования категориальных переменных
import numpy as np  # Для математических операций и работы с массивами
import matplotlib.pyplot as plt  # Для построения графиков

# Загрузка и предварительная обработка данных

# Загрузка данных из CSV-файла
data = pd.read_csv('cirrhosis.csv') 

# Преобразуем целевую переменную (статус) в числовой формат
data['Status'] = data['Status'].map({'D': 0, 'C': 1, 'CL': 2})  # Преобразование классов в числа для использования в модели

# Обработка категориальных признаков
categorical_cols = data.select_dtypes(include=['object']).columns  # Определяем все столбцы с категориальными данными
label_encoders = {}  # Словарь для хранения LabelEncoder для каждого столбца
for col in categorical_cols:
    le = LabelEncoder()  # Создаем объект LabelEncoder для кодирования категориального столбца
    data[col] = le.fit_transform(data[col].astype(str))  # Кодируем значения столбца в числовые значения
    label_encoders[col] = le  # Сохраняем LabelEncoder для возможного декодирования в будущем

# Удаление ненужных столбцов
X = data.drop(columns=['ID', 'Status'])  # Признаки (features), исключая ID и целевую переменную Status
y = data['Status']  # Целевая переменная (статус: смерть, выход из исследования или трансплантация)

# Разделение данных на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)  # 70% для обучения, 30% для теста

# Построение и визуализация дерева решений (Задание 2.1)

# Задание гиперпараметров дерева решений
max_depth = 5  # Ограничение глубины дерева (максимум 5 уровней)
max_features = 10  # Ограничение числа признаков, используемых для разбиения в каждом узле

# Инициализация и обучение дерева решений
clf = DecisionTreeClassifier(max_depth=max_depth, max_features=max_features, random_state=42)  # Создаем классификатор
clf.fit(X_train, y_train)  # Обучаем модель на обучающей выборке

# Визуализация дерева решений
plt.figure(figsize=(20, 10))  # Задаем размер графика
plot_tree(clf, feature_names=X.columns, class_names=['D', 'C', 'CL'], filled=True)  # Построение графика дерева
plt.title(f"Decision Tree (max_depth={max_depth}, max_features={max_features})")  # Заголовок графика
plt.savefig('decision_tree.png')  # Сохранение изображения дерева в файл
plt.show()  # Отображение графика

# Вывод результатов
print("Визуализация дерева решений с глубиной 5 и использованием до 10 признаков.")
print("Дерево эффективно разделяет данные на три класса: D (смерть), C (выбыл из исследования) и CL (трансплантация).")
print("Разбиение узлов происходит на основе значений признаков, таких как 'N_Days', минимизируя критерий Gini для улучшения классификации.\n")

# Анализ влияния глубины дерева на качество модели (Задание 2.2)

# Проверка качества модели для различных значений max_depth
depths = range(1, 21)  # Диапазон значений глубины дерева от 1 до 20
mse_depths = []  # Список для хранения MSE (среднеквадратической ошибки) для каждой глубины

for depth in depths:  # Цикл по каждому значению глубины
    clf = DecisionTreeClassifier(max_depth=depth, random_state=42)  # Инициализация дерева с текущей глубиной
    scores = cross_val_score(clf, X_train, y_train, scoring='neg_mean_squared_error', cv=5)  # 5-кратная кросс-валидация
    mse_depths.append(-scores.mean())  # Сохраняем MSE (отрицательный результат кросс-валидации)

# Визуализация зависимости MSE от глубины дерева
plt.figure(figsize=(10, 6))  # Размер графика
plt.plot(depths, mse_depths, marker='o', label='MSE')  # Линия с маркерами для значений MSE
plt.xlabel('Max Depth')  # Метка оси X
plt.ylabel('Mean Squared Error (MSE)')  # Метка оси Y
plt.title('MSE vs Max Depth')  # Заголовок графика
plt.legend()  # Легенда графика
plt.grid()  # Включение сетки
plt.savefig('mse_vs_max_depth.png')  # Сохранение графика в файл
plt.show()  # Отображение графика

# Вывод результатов
print("График зависимости MSE от глубины дерева показывает, что ошибка уменьшается с увеличением глубины.")
print("После определенного значения (глубина около 12) MSE стабилизируется, что свидетельствует об оптимальной глубине для минимизации переобучения.\n")

print("Max Depth (Максимальная глубина дерева) представляет максимальную глубину дерева решений, которая определяет, насколько глубоко модель может разветвлять дерево для классификации данных. Значения изменяются от 1 до 20 (по заданному диапазону).\n")

print("Mean Squared Error (MSE) Среднеквадратическая ошибка, показывающая, насколько прогнозы модели отклоняются от истинных значений. Чем меньше значение, тем точнее модель.\n")

# Анализ влияния количества признаков на качество модели (Задание 2.3)

# Проверка качества модели для различных значений max_features
features = range(1, X.shape[1] + 1)  # Диапазон от 1 до числа признаков в данных
mse_features = []  # Список для хранения MSE для каждого значения max_features

for feature in features:  # Цикл по каждому значению max_features
    clf = DecisionTreeClassifier(max_features=feature, random_state=42)  # Инициализация дерева с текущим max_features
    scores = cross_val_score(clf, X_train, y_train, scoring='neg_mean_squared_error', cv=5)  # 5-кратная кросс-валидация
    mse_features.append(-scores.mean())  # Сохраняем MSE

# Визуализация зависимости MSE от количества признаков
plt.figure(figsize=(10, 6))  # Размер графика
plt.plot(features, mse_features, marker='o', label='MSE')  # Линия с маркерами
plt.xlabel('Max Features')  # Метка оси X
plt.ylabel('Mean Squared Error (MSE)')  # Метка оси Y
plt.title('MSE vs Max Features')  # Заголовок графика
plt.legend()  # Легенда графика
plt.grid()  # Сетка
plt.savefig('mse_vs_max_features.png')  # Сохранение графика
plt.show()  # Отображение графика

# Вывод результатов
print("График зависимости MSE от количества признаков показывает, что ошибка минимальна при использовании определенного числа признаков (около 11).")
print("Дальнейшее увеличение числа признаков может привести к ухудшению обобщающей способности модели.\n")

print("Max Features (Максимальное количество признаков) Количество признаков, которые модель может учитывать при разбиении данных в каждом узле. Значения изменяются от 1 до количества всех признаков в данных (в данном случае до 18).\n")

print("Mean Squared Error (MSE) Среднеквадратическая ошибка, аналогичная описанной выше. Показатель указывает на точность модели в зависимости от количества используемых признаков.\n")

# Оптимизация гиперпараметров (Задание 2.4)

# Определение оптимальных параметров
optimal_depth = depths[np.argmin(mse_depths)]  # Глубина с минимальным значением MSE
optimal_features = features[np.argmin(mse_features)]  # Количество признаков с минимальным MSE

# Вывод оптимальных параметров
print(f"Optimal Max Depth: {optimal_depth}")  # Оптимальная глубина дерева
print(f"Optimal Max Features: {optimal_features}\n")  # Оптимальное количество признаков

# Обучение модели с оптимальными параметрами
clf_optimal = DecisionTreeClassifier(max_depth=optimal_depth, max_features=optimal_features, random_state=42)  # Создаем модель
clf_optimal.fit(X_train, y_train)  # Обучаем модель

# Визуализация оптимального дерева решений (Задание 2.5)

# Построение дерева с оптимальными параметрами
plt.figure(figsize=(20, 10))  # Размер графика
plot_tree(clf_optimal, feature_names=X.columns, class_names=['D', 'C', 'CL'], filled=True)  # Построение графика дерева
plt.title(f"Optimal Decision Tree (max_depth={optimal_depth}, max_features={optimal_features})")  # Заголовок графика
plt.savefig('optimal_decision_tree.png')  # Сохранение графика дерева
plt.show()  # Отображение графика

# Вывод результатов
print("Визуализация оптимального дерева решений с параметрами: глубина 12 и количество признаков 11.")
print("Модель минимизирует ошибку и обеспечивает более точную классификацию состояний выживаемости.\n")
```
