**В качестве набора ресурсов, буду использовать winequality-red.csv и winequality-white.csv**

**Описание набора**

**Включены два набора данных, связанных с образцами красного и белого вина vinho verde с севера Португалии. Цель состоит в том, чтобы смоделировать качество вина на основе физико-химических испытаний**

**Два набора данных относятся к красным и белым сортам португальского вина «Винью Верде». Для получения более подробной информации обратитесь по адресу: http://www.vinhoverde.pt/en/ или к источнику [Cortez et al., 2009]. Из-за проблем с конфиденциальностью и логистикой доступны только физико-химические (входные) и сенсорные (выходные) переменные (например, нет данных о сортах винограда, марке вина, цене продажи и т. д.). Эти наборы данных можно рассматривать как задачи классификации или регрессии. Классы упорядочены и не сбалансированы (например, нормальных вин намного больше, чем отличных или плохих). Алгоритмы обнаружения выбросов могут быть использованы для обнаружения нескольких отличных или плохих вин. Кроме того, мы не уверены, что все входные переменные релевантны. Поэтому было бы интересно протестировать методы выбора функций.**

**Переменные таблицы:**

**Входные переменные (на основе физико-химических испытаний):**

 **1. fixed_acidity — общая кислотность**
 
 **2. volatile_acidity — летучая кислотность**
 
 **3. citric_acid — лимонная кислота**
 
 **4. residual_sugar — остаточный сахар**
 
 **5. chlorides — хлориды**
 
 **6. free_sulfur_dioxide — свободный диоксид серы**
 
 **7. total_sulfur_dioxide — общий диоксид серы**
 
 **8. density — плотность**
 
 **9. pH — это водородный показатель среды, или показатель кислотности**
  
 **10. sulphates — сульфаты**
  
 **11. alcohol — алкоголь**
  
 **Выходная переменная (на основе сенсорных данных):** 
 
 **12. quality — качество (оценка от 0 до 10)**

**Для выполнения 2 лабораторной работы необходимо:**

Проведите первичный анализ данных. Особое внимание следует 
уделить графическому представлению распрееделений признаков, 
визуализации взаимосвзей, позволяющие судить о наборе данных. Построение 
графиков желательно произвести по нескольким проекциям. При анализе 
данных использовать как можно более разнообразные типы графиков.

```python
# Импорт необходимых библиотек
import pandas as pd  # Для работы с данными в табличном формате (DataFrame)
import numpy as np  # Для выполнения математических операций и работы с массивами
import matplotlib.pyplot as plt  # Для построения графиков и визуализации данных
import seaborn as sns  # Для создания более красивых и информативных графиков на базе matplotlib

# Установим стиль графиков (белый фон с сеткой)
sns.set(style="whitegrid")  # Применяем стиль "whitegrid" для улучшения визуализации

# Загрузка данных из CSV файлов для красного и белого вина
red_wine = pd.read_csv('winequality-red.csv', sep=';')  # Загрузка данных о красном вине
white_wine = pd.read_csv('winequality-white.csv', sep=';')  # Загрузка данных о белом вине

# Добавим колонку "wine_type" для различия типов вина
# Это необходимо, чтобы позже различать два типа вина в графиках и анализах
red_wine['wine_type'] = 'Red'  # Для красного вина добавляем метку 'Red'
white_wine['wine_type'] = 'White'  # Для белого вина добавляем метку 'White'

# Объединим данные о красном и белом вине в один DataFrame для удобства анализа
wine_data = pd.concat([red_wine, white_wine], axis=0)  # Соединяем данные по обеим винам в один набор

# Первичный анализ данных
# Выводим информацию о наборе данных: количество строк, колонок и типы данных
print("Общая информация о данных:")
print(wine_data.info())  # Информация о данных: размеры (строки, столбцы) и типы данных

# Основные статистики по числовым колонкам: среднее, стандартное отклонение, минимальные и максимальные значения
print("\nОсновные статистики:")
print(wine_data.describe())  # Статистическое описание числовых признаков: min, max, mean и т.д.

# Проверка на пропуски в данных
# Проверяем, если есть отсутствующие значения в столбцах
print("\nПропущенные значения:")
print(wine_data.isnull().sum())  # Проверка на наличие пропущенных значений в каждом столбце

# Преобразуем данные для корректной работы корреляционной матрицы
# Кодируем тип вина как числа: 'Red' -> 0, 'White' -> 1
wine_data_encoded = wine_data.copy()  # Создаем копию данных, чтобы не изменять оригинал
wine_data_encoded['wine_type'] = wine_data_encoded['wine_type'].map({'Red': 0, 'White': 1})  # Кодируем тип вина

# 1. Гистограмма распределения качества вина
# Строим гистограмму для анализа распределения оценок качества
plt.figure(figsize=(8, 6))  # Задаем размер графика
sns.histplot(data=wine_data, x="quality", hue="wine_type", multiple="dodge", palette="husl", bins=10)  # Строим гистограмму с раздельными цветами для типов вин
plt.title("Распределение качества вина")  # Заголовок графика
plt.xlabel("Качество")  # Подпись оси X
plt.ylabel("Частота")  # Подпись оси Y
plt.legend(title="Тип вина", labels=["Красное", "Белое"])  # Легенда для графика
plt.show()  # Отображаем график

# Вывод 1: Анализ распределения качества вина
print("Вывод 1:")
print("- Красное вино получает более высокие оценки, чем ,белое , что может свидетельствовать о его лучшем общем качестве в наборе данных.")
print("- Белое вино чаще оценивается средними баллами, что может быть связано с его характеристиками или меньшей вариативностью качества.\n")

# 2. Боксплот для сравнения содержания алкоголя
# Строим боксплот для анализа распределения содержания алкоголя по типу вина
plt.figure(figsize=(8, 6))  # Задаем размер графика
sns.boxplot(data=wine_data, x="wine_type", y="alcohol", hue="wine_type", dodge=False, palette={"Red": "red", "White": "blue"})  # Боксплот для сравнения по типам вина
plt.title("Сравнение содержания алкоголя по типу вина")  # Заголовок графика
plt.xlabel("Тип вина")  # Подпись оси X
plt.ylabel("Содержание алкоголя (%)")  # Подпись оси Y
plt.legend(title="Тип вина", labels=["Красное", "Белое"])  # Легенда для графика
plt.show()  # Отображаем график

# Вывод 2: Анализ содержания алкоголя
print("Вывод 2:")
print("- Содержание алкоголя в белом вине в среднем выше, чем в красном.")
print("- Диапазон значений алкоголя у белого вина шире, чем у красного, что указывает на более разнообразный ассортимент.")
print("- У красного вина меньше выбросов по содержанию алкоголя, что может говорить о стандартизированном процессе производства.\n")

# 3. Корреляционная матрица
# Строим корреляционную матрицу для анализа взаимосвязей между признаками
plt.figure(figsize=(10, 8))  # Задаем размер графика
corr_matrix = wine_data_encoded.corr()  # Вычисляем корреляцию между признаками
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)  # Строим тепловую карту для корреляций
plt.title("Корреляционная матрица признаков")  # Заголовок графика
plt.show()  # Отображаем график

# Вывод 3: Анализ корреляции
print("Вывод 3:")
print("- Сильная положительная корреляция между содержанием алкоголя и качеством вина указывает на то, что алкоголь может быть важным фактором, влияющим на оценку качества.")
print("- Плотность имеет обратную корреляцию с содержанием алкоголя, особенно у белого вина.")
print("- Остальные корреляции менее выражены, что может указывать на слабое влияние остальных признаков на качество.\n")

# 4. Распределение плотности (density)
# Строим график распределения плотности по типам вина для анализа различий
plt.figure(figsize=(8, 6))  # Задаем размер графика
sns.kdeplot(data=wine_data, x="density", hue="wine_type", fill=True, common_norm=False, palette={"Red": "red", "White": "blue"})  # График плотности для типов вина
plt.title("Распределение плотности по типам вина")  # Заголовок графика
plt.xlabel("Плотность")  # Подпись оси X
plt.ylabel("Плотность распределения")  # Подпись оси Y
plt.show()  # Отображаем график

# Вывод 4: Анализ плотности
print("Вывод 4:")
print("- Красное вино имеет более высокую плотность, чем белое, что может быть связано с использованием других сортов винограда или особенностями технологии производства.")
print("- Белое вино чаще встречается с низкой плотностью, что связано с его легкостью и свежестью.\n")

# 5. Парный график для нескольких признаков
# Строим парный график для изучения взаимосвязей между признаками: alcohol, density, pH, quality
sns.pairplot(wine_data, vars=['alcohol', 'density', 'pH', 'quality'], hue='wine_type', palette={"Red": "red", "White": "blue"}, diag_kind="kde")  # Парный график для нескольких признаков
plt.suptitle("Парные графики для основных признаков", y=1.02)  # Заголовок графика
plt.show()  # Отображаем график

# Вывод 5: Анализ парных графиков
print("Вывод 5:")
print("- Содержание алкоголя имеет явную положительную связь с качеством, особенно у белого вина.")
print("- Плотность и качество связаны отрицательно, что подтверждает более высокие оценки для легких вин.")
print("- pH имеет слабую связь с качеством, однако у красного вина наблюдаются небольшие отличия в распределении.")
print("- Парные графики показывают, что белое и красное вино имеют заметные различия в признаках, что подтверждает их разную природу.\n")
```
